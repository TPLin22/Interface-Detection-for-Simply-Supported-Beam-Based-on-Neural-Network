```python
import numpy as np
import pandas as pd
import random
import time

start_time = time.time()


# Placeholder for the formula
def f(x, l1, l2, Omega, EI1, EI2):
    x=int(x)
    l = l1+l2
    a = l1
    q = Omega
    if x<=l1:
        #v_x=dv0*x+(Omega*l1**4)/(24*EI1)+Omega*l1**3*x*(1/(6*EI1)-1/(6*EI2))+(Omega*x**4)/(24*EI2)-Omega*l1**4*(1/(6*EI1)-1/(8*EI2))
        v_x=(q*l*(x**3)/12-q*(x**4)/24+x*(q*(a**3)/3-q*l*(a**2)/4-q*(a**4)/8/l))/EI1+(q*l*(a**2)/4-q*(a**3)/3+q*(a**4)/8/l-q*(l**3)/24)*x/EI2
    else:
        v_x=(q*l*(x**3)/12-q*(x**4)/24+x*(q*(a**4)/8/l-q*(a**3)/6-q*(l**3)/24)+q*l*(a**3)/6-q*(a**4)/8)/EI2+(x*(q*(a**3)/6-q*(a**4)/8/l)+q*(a**4)/8-q*l*(a**3)/6)/EI1
    return -v_x  

data = []
num_rows = 100000

for i in range(num_rows):
    l1 = random.randint(0,1001)
    l2, Omega= 1000-l1, 100
    #EI1, EI2 = 100000000000 + EI1_error, 200000000000 + EI2_error
    EI1 = random.uniform(100000000000-20000000000,100000000000+20000000000)
    EI2 = random.uniform(200000000000-20000000000,200000000000+20000000000)
    row = [l1] + [EI1] + [EI2] + [f(x, l1, l2, Omega, EI1, EI2) for x in range(0,1001)]
    data.append(row)
    if i % 1000 == 0:
        elapsed_time = time.time() - start_time
        print("Step: {} - Elapsed Time: {:.2f} seconds".format(i, elapsed_time))
        
        

# Create DataFrame
df = pd.DataFrame(data, columns=["boundary"] + ["EI1"] + ["EI2"] + [f"x_{i}" for i in range(0,1001)])

# Save to CSV file
csv_filename = "./Random_E1E2_million_of_1000_Omega=100_EI1=1e11_EI2=2e11_phr.csv"
df.to_csv(csv_filename, index=False)
```
